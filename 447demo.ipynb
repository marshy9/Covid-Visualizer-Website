```python
import sqlite3 as lite
import pandas as pd
from pandas import DataFrame
import sys
from os import system, name 

# decides if tables need to be created or not
def init(con, table):
    
    no = ["No", "NO", "no", "n", "N"]

    db_made = input("Have you created the database yet: ") # only need to create tables and load csv files once
    
    if db_made in no:

        execute_command(con, table)

        load(con)

# creates the relational tables
def execute_command(con, sql_stmt):

    #create cursor and execute sql statement inputted
    cur = con.cursor()
    cur.execute(sql_stmt)

    # save transaction
    con.commit()

# loads and reads in data from csv files
def load(con):

    # uses pandas package to read in csv files to a dataframe 
    # and then converts to SQLite tables

    # to test with different data replace the arguments within the 'pd.read_csv' methods
    path_covid = r"https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"
    read_covid = pd.read_csv(path_covid)
    read_covid.to_sql('COVID', con, if_exists = "append", index = False)
    
    path_prison = r"C:\Users\andre\Desktop\PrisonData.csv"
    read_prison = pd.read_csv(path_prison)
    read_prison.to_sql('PRISON', con, if_exists = "append", index = False)
    
def query(cur, con):
    # section where SQLite queries are ran

    try:
        myStmt = "go"
        while myStmt != "quit":
            system('cls')
            myStmt = input('Enter SQL Statement: ')
            if myStmt != "quit":
                cur.execute(myStmt)
                data = cur.fetchall()
                print("The results are: ")
                for rec in data:
                    for field in rec:
                        print(field,"\t")
                    print()
                print()
                myWait = input('Press enter to continue: ')
            con.commit()
        con.close()

    except lite.Error as e:

        print("Error %s:" % e.args[0])
        #sys.exit(1)


def main():

    con = None

    # file name
    database = "projDB.db"

    # create connection to our database
    con = lite.connect(database)

    # create cursor for SQLite tasks
    cur = con.cursor()

    drop_table_covid = '''DROP TABLE IF EXISTS COVID'''
    
    execute_command(con, drop_table_covid)
    
    sql_create_covid = ''' CREATE TABLE IF NOT EXISTS COVID
                            ( 
                            [id] INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
                            [date] DATE, 
                            [county] VARCHAR(30),
                            [state] VARCHAR(20),
                            [fips] INTEGER,
                            [cases] INTEGER,
                            [deaths] INTEGER
                            )'''

    init(con, sql_create_covid)
    
    
    
    drop_table_prison = '''DROP TABLE IF EXISTS PRISON'''
    
    execute_command(con, drop_table_prison)
    
    sql_create_prison = ''' CREATE TABLE IF NOT EXISTS PRISON
                            (
                             [id] INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
                             [Date] DATE,
                             [Facility Type] varchar(50),
                             [State] varchar(50),
                             [Canonical Facility Name] varchar(50),
                             [Pop Tested] int,
                             [Pop Tested Positive] int,
                             [Pop Tested Negative] int,
                             [Pop Deaths] int,
                             [Pop Recovered] int,
                             [Staff Tested] int,
                             [Staff Tested Positive] int,
                             [Staff Tested Negative] int,
                             [Staff Deaths] int,
                             [Staff Recovered] int,
                             [Source] varchar(200),
                             [Compilation] varchar(100),
                             [Notes] varchar(50)
                             )'''
        
    
    init(con, sql_create_prison)

    query(cur, con)

if __name__ == "__main__":
    main()
```
